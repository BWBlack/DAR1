```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
seed_file_3d = "/Users/birdy/Documents/eaps_research/gcm_analysis/gcm_data/darwin_weekly_seasonal/3d.nc"
seed_file_temp = "/Users/birdy/Documents/eaps_research/gcm_analysis/gcm_data/darwin_weekly_seasonal/tave.nc"
seed_file_par = "/Users/birdy/Documents/eaps_research/gcm_analysis/gcm_data/darwin_weekly_seasonal/par.nc"
seed_ds_3d = xr.open_dataset(seed_file_3d)
seed_ds_temp = xr.open_dataset(seed_file_temp)
seed_ds_par = xr.open_dataset(seed_file_par)
```

```python
seed_ds_temp["Ttave"].isel(T=5, Z=0).plot()
```

```python
seed_ds_3d_avg = seed_ds_3d.mean(dim='T', keepdims=True)
seed_ds_temp_avg = seed_ds_temp.mean(dim='T', keepdims=True)
seed_ds_par_avg = seed_ds_par.mean(dim='T', keepdims=True)
```

```python
# write to file 
seed_file_3d_out = "/Users/birdy/Documents/eaps_research/gcm_analysis/gcm_data/darwin_yearly_avg/3d.nc"
seed_file_temp_out = "/Users/birdy/Documents/eaps_research/gcm_analysis/gcm_data/darwin_yearly_avg/tave.nc"
seed_file_par_out = "/Users/birdy/Documents/eaps_research/gcm_analysis/gcm_data/darwin_yearly_avg/par.nc"

seed_ds_3d_avg.to_netcdf(seed_file_3d_out)
seed_ds_temp_avg.to_netcdf(seed_file_temp_out)
seed_ds_par_avg.to_netcdf(seed_file_par_out)
```
