```python
# sum biomass

# dset: xarray dataset or dataArray 
# variables: list of variables that exist in dset to be summed 
# new_name: what to call the new variable
# returns the dset with a new variable created with new_name and attributes
def ds_sum_variables(dset, variables, new_name, new_units="", new_description=""):
    c_total = dset[variables[0]]
    for var in variables[1:]:
        c_next = dset[var]
        c_total = c_total + c_next
    c_total = c_total.assign_attrs({'units':new_units, 'description': new_description})
    dset = dset.assign({new_name:c_total})
    return dset

biomass_vars = []
for i in range(21,70):
    biomass_vars.append(f"TRAC{i}")
```

Load up the run WITHOUT extra nutrients

```python

def make_graphs(file_list):
    # open control
    ds_control = xr.open_mfdataset(file_list[0])
    ds_control = ds_sum_variables(ds_control, biomass_vars, "total_biomass", new_units="mmol C /")
    
    # open no3 added
    ds_no3 = xr.open_mfdataset(file_list[1])
    ds_no3 = ds_sum_variables(ds_no3, biomass_vars, "total_biomass", new_units="mmol C /")

    # open nh4 added 
    ds_nh4 = xr.open_mfdataset(file_list[2])
    ds_nh4 = ds_sum_variables(ds_nh4, biomass_vars, "total_biomass", new_units="mmol C /")
    
    fig, ax = plt.subplots()
    l1, = ds_control["total_biomass"].plot(ax=ax, color='g', ls="--", lw=3)
    l2, = ds_no3["total_biomass"].plot(ax=ax, color='skyblue', lw=3)
    #ds_nh4["total_biomass"].plot(ax=ax)
    ax.set_title("Incubation Timeseries", fontsize=16)
    ax.legend([l1, l2], ["control", "added 10x NO3"])
    ax.set_xticklabels(range(7))
    ax.set_xlabel("Days")
    fig.set_size_inches(5,5)
    plt.show()
    
    fig, ax = plt.subplots()
    l1, = ds_control["TRAC02"].plot(ax=ax, color='g', ls="--", lw=3)
    l2, = ds_no3["TRAC02"].plot(ax=ax, color='skyblue', lw=3)
    #ds_nh4["total_biomass"].plot(ax=ax)
    ax.set_title("Total NO3", fontsize=16)
    ax.legend([l1, l2], ["control", "added 10x NO3"])
    ax.set_xticklabels(range(7))
    ax.set_xlabel("Days")
    fig.set_size_inches(5,5)
    plt.show()
    
    return ds_control, ds_no3, ds_nh4

```
