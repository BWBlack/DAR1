# Propocalypse - with Light

```python
file = "/Users/birdy/Documents/eaps_research/darwin3/verification/dar_one_config/run/gp-1/run/ecco_gud_20221205_0001/3d*"
ds = xr.open_mfdataset(file)
ds
```

```python
ds["TRAC21"].isel(T=slice(5,6)).mean(dim="T").plot()
```

```python
# difference between adding light and no light 
no_light_da = ds_no_light["TRAC21"].isel(T=slice(-2,-1)).mean(dim="T")
light_da = ds["TRAC21"].isel(T=slice(5,6)).mean(dim="T")
diff_da = no_light_da - light_da
```

```python
x_axis = seed_ds_3d["T"].data
y_axis = seed_ds_3d.isel(Y=slice(100, 137))["Y"].data

diff_da.assign_coords({"Y":y_axis})
diff_da.assign_coords({"X":x_axis})
```

```python
pretty_plot_lat_over_time(diff_da)
```

```python
def pretty_plot_lat_over_time(da, title=""):
    x_axis = seed_ds_3d["T"].data
    y_axis = seed_ds_3d.isel(Y=slice(100, 137))["Y"].data

    my_da = da.assign_coords({"Y":y_axis})
    my_da = my_da.assign_coords({"X":x_axis})

    fig, ax = plt.subplots()
    my_da.plot(ax=ax)
    ax.set_xlabel("Time")
    ax.set_ylabel("Latitude")
    ax.title
```
