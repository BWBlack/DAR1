# Getting Large-Model Propocalypse Path

```python
file = "/Users/birdy/Documents/eaps_research/gcm_analysis/gcm_data/darwin_weekly_seasonal/3d.nc"
tracer_ds = xr.open_dataset(file)
tracer_ds = tracer_ds.rename({'Zmd000023':'Z'})
tracer_ds
```

```python
lon = 205.5
z = 0

# contour plot the propocalyse over gradients location  
fig, ax = plt.subplots()
n_contour_colors = 2 
levels=[.001, 1]
pro_contour = tracer_ds["TRAC21"].isel(Z=z).sel(X=lon, Y=slice(20,60)).T.plot.contourf(ax=ax, levels=[.001, 1])
ax.set_title("Location of Propocalypse over Time")
plt.show()
```

```python
# get the path of the propocalypse
p1 = pro_contour.collections[0].get_paths()[1]
p1_coords = p1.vertices

propocalypse_x = []
propocalypse_y = []

for coord in p1_coords:
    # remove top 
    if coord[1] == 59.5:
        continue
    # remove sides 
    if coord[0] == 14555.0 or coord[0] == 14198.0:
        continue
    propocalypse_x.append(coord[0])
    propocalypse_y.append(coord[1]) 
    
propocalypse_x = np.array(propocalypse_x)
propocalypse_y = np.array(propocalypse_y)

# convert time back into datatime 
propocalypse_time = np.array(list(map(lambda x: np.datetime64(int(x), 'D'), propocalypse_x)))

# sort 
temp = np.array(sorted(list(zip(propocalypse_y, propocalypse_time)), key=lambda x: x[1]))
propocalypse_lat = temp[:,0]
propocalypse_time = temp[:,1]
```

```python
# show the location of the points is correct  
fig, ax = plt.subplots()
n_contour_colors = 2 
tracer_ds["TRAC21"].isel(Z=z).sel(X=lon, Y=slice(20,60)).T.plot.contourf(ax=ax, levels = levels)
ax.set_title("Location of Propocalypse over Time")
ax.scatter(propocalypse_time, propocalypse_lat, c='r')
plt.show() 
```
