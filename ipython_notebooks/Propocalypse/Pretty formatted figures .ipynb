# Global Model vs Steady State 

Steady state model is seeded with proper light, temperature, biomass, and nutrients 

```python
# NO CBAR SCALING 
fig, [[ax1, ax2], [ax3, ax4], [ax5, ax6]] = plt.subplots(3,2)

#pro
seed_ds_3d["TRAC21"].isel(Z=z).sel(X=x, Y=y).T.plot(ax=ax1)
ax1.set_title("Global Model: Pro")
pretty_plot_lat_over_time(ax2, tracer_ds["TRAC21"])
ax2.set_title("Steady State: Pro")

#syn
seed_ds_3d["TRAC22"].isel(Z=z).sel(X=x, Y=y).T.plot(ax=ax3)
ax3.set_title("Global Model: Syn")
pretty_plot_lat_over_time(ax4, tracer_ds["TRAC22"])
ax4.set_title("Steady State: Syn")

#het 
seed_hets = seed_ds_3d["TRAC68"] + seed_ds_3d["TRAC69"] + seed_ds_3d["TRAC70"]
ss_hets = tracer_ds["TRAC68"] + tracer_ds["TRAC69"] + tracer_ds["TRAC70"]  
seed_hets.isel(Z=z).sel(X=x, Y=y).T.plot(ax=ax5)
ax5.set_title("Global Model: Heterotrophs")
pretty_plot_lat_over_time(ax6, ss_hets)
ax6.set_title("Steady State: Heterotrophs")

for ax in [ax1, ax2, ax3, ax4]:
    ax.get_xaxis().set_ticklabels([])


fig.set_size_inches(10,10)
fig.tight_layout()
plt.show()
```
